```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import io
%pylab inline
#replacing commas and string in order to convert it to floats
def cleanNumbers (s):
    output = s.replace (',' , '')
    return float(output)

#reading the statistical file
death = pd.read_csv("NCHS_-_Leading_Causes_of_Death__United_States.csv")
#applies all values in Deaths to the function cleanNumbers
death['Death_num'] = death['Deaths'].apply(cleanNumbers)

#summing up death number by state and renaming the column to total deaths
c = death.groupby('State')['Death_num'].mean().reset_index(name='Total Deaths')
#sorting values based on total deaths
c = c.sort_values('Total Deaths', ascending=False)
#selecting all rows where state is not the United State
c = c[c.State != 'United States']
#adjusting figure size 
plt.figure(figsize(40,10))
# x-axis is state and y-axis is total deaths
plt.bar(c['State'],c['Total Deaths'])
#saving the figure as a separate file
plt.savefig('figure.pdf')
print(c)
#print(death)

#find the change in deaths for a certain cause 
cn = input("Enter cause of death you wish to research")
cid = death.query('Cause Name == cn')
#compare the changes in death for a certain cause 
#possible idea: store in variables the total deaths numbers for rows containing the year 1999 & 2016 
#Subtract variables to get changes in deaths rates
f =cid.groupby('Total Deaths').mean()
#organize table/data
x = f[]
y = f[]
plt.scatter(x,y)
plt.xlabel("States")
plt.ylabel("Changes in Number of Deaths from 1999 to 2016")




```
